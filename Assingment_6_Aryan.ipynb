## --- 1. parse both the original files from the sct and original directory and combine them to make merged_df as ground truth ---

```python
import os
import pandas as pd

def merge_original_and_sct(original_file_path, sct_file_path):
  '''
      Reads original and sct files, parses them into dataframes,
      and merges them on the 'label' column, dropping the 'value_x' column.
  '''
  # Logic from original_to_dataframe
  with open(original_file_path, "r") as file:
    texts = file.read()
    texts = texts.split("\n")
    texts = [i.split("\t") for i in texts]
    texts = [i for i in texts if i != [''] and i[0][0] != '#']

  original_data = []
  for value in texts:
    my_dict = {}
    my_dict["label"] = "T" + value[0]
    my_dict["identifier"] = value[1].split(" ")[0]
    my_dict["value"] = value[2]
    original_data.append(my_dict)

  original_df = pd.DataFrame(original_data, columns=["label", "identifier", "value"])

  # Logic from sct_to_dataframe
  sct_dict_list = []
  with open(sct_file_path, "r") as file:
    texts = file.read()
    texts = texts.split("\n")
    texts = [i.split("\t") for i in texts][:-1]

  for text in texts:
    data = {}
    data["label"] = text[0]
    data["value"] = text[2]
    data["sct_name"] = text[1].split(" | ")[1]
    sct_dict_list.append(data)

  sct_df = pd.DataFrame(sct_dict_list)


  merged_df = pd.merge(original_df, sct_df, on='label', how='inner').drop('value_x', axis=1)
  return merged_df

# Use the new combined function for all 5 pairs
merged_text1 = merge_original_and_sct('/content/drive/MyDrive/cadec/original/ARTHROTEC.1.ann', '/content/drive/MyDrive/cadec/sct/ARTHROTEC.1.ann')
merged_text2 = merge_original_and_sct('/content/drive/MyDrive/cadec/original/ARTHROTEC.2.ann', '/content/drive/MyDrive/cadec/sct/ARTHROTEC.2.ann')
merged_text3 = merge_original_and_sct('/content/drive/MyDrive/cadec/original/ARTHROTEC.3.ann', '/content/drive/MyDrive/cadec/sct/ARTHROTEC.3.ann')
merged_text4 = merge_original_and_sct('/content/drive/MyDrive/cadec/original/ARTHROTEC.4.ann', '/content/drive/MyDrive/cadec/sct/ARTHROTEC.4.ann')
merged_text5 = merge_original_and_sct('/content/drive/MyDrive/cadec/original/ARTHROTEC.5.ann', '/content/drive/MyDrive/cadec/sct/ARTHROTEC.5.ann')

# Display one of the merged dataframes as an example
display(merged_text1)
display(merged_text2)
display(merged_text3)
display(merged_text4)
display(merged_text5)
```

## --- 2. parse the model outputs from original2 folder ---

```python
def parse_model_otput_ADR_to_df(file_path):
  pred_textdf = pd.read_csv(file_path)
  pred_textdf = pred_textdf[pred_textdf["identifier"] == "ADR"]
  return pred_textdf

pred_text1 = parse_model_otput_ADR_to_df("/content/drive/MyDrive/cadec/original2/ARTHROTEC.1.txt")
pred_text2 = parse_model_otput_ADR_to_df("/content/drive/MyDrive/cadec/original2/ARTHROTEC.2.txt")
pred_text3 = parse_model_otput_ADR_to_df("/content/drive/MyDrive/cadec/original2/ARTHROTEC.3.txt")
pred_text4 = parse_model_otput_ADR_to_df("/content/drive/MyDrive/cadec/original2/ARTHROTEC.4.txt")
pred_text5 = parse_model_otput_ADR_to_df("/content/drive/MyDrive/cadec/original2/ARTHROTEC.5.txt")

pred_text1
```

## --- 3. trying to label sct_names for each text using approximate string match ---

```python
from fuzzywuzzy import fuzz
import pandas as pd

'''
    For each row in the pred_text go to all rows in merged_df and try to match string.
    if the score is good enough it will be labeled as the sct_name
    makes a combined_pred_text_df for ease of model evaluation

'''
pred_text_dfs = [pred_text1, pred_text2, pred_text3, pred_text4, pred_text5]
merged_dfs = [merged_text1, merged_text2, merged_text3, merged_text4, merged_text5]
all_updated_pred_dfs = []

for pred_df in pred_text_dfs:
    sct_names_for_df = []
    for index, row in pred_df.iterrows():
        pred_text_value = row['text']
        best_match_sct_name = None
        highest_ratio = 0
        for merged_df in merged_dfs:
            for merged_index, merged_row in merged_df.iterrows():
                merged_text_value = merged_row['value_y']
                similarity_ratio = fuzz.ratio(pred_text_value, merged_text_value)
                # Consider a match if similarity is above 80
                if similarity_ratio > 80 and similarity_ratio > highest_ratio:
                    highest_ratio = similarity_ratio
                    best_match_sct_name = merged_row['sct_name']
        sct_names_for_df.append(best_match_sct_name)
    pred_df['sct_name'] = sct_names_for_df
    all_updated_pred_dfs.append(pred_df)


combined_pred_text_df = pd.concat(all_updated_pred_dfs, ignore_index=True)

display(combined_pred_text_df)
```

## --- 4. trying to label sct_name for each text using embedding HF model ---

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd


model = SentenceTransformer('all-MiniLM-L6-v2')

pred_text_dfs = [pred_text1, pred_text2, pred_text3, pred_text4, pred_text5]
merged_dfs = [merged_text1, merged_text2, merged_text3, merged_text4, merged_text5]
all_updated_pred_dfs = []

# Generate embeddings for merged_df
for df in merged_dfs:
    df['value_y_embedding'] = df['value_y'].apply(lambda x: model.encode(x))

# Generate embeddings for pred text data
for df in pred_text_dfs:
    df['text_embedding'] = df['text'].apply(lambda x: model.encode(x))


def calculate_similarity(single_embedding, embedding_list):
  """
      Calculates cosine similarity between a single embedding and a list of embeddings.

  """

  single_embedding_2d = single_embedding.reshape(1, -1)
  similarity_scores = cosine_similarity(single_embedding_2d, embedding_list)
  return similarity_scores[0]


for pred_df in pred_text_dfs:
    sct_names_for_df = []
    similarity_scores_for_df = []
    for index, row in pred_df.iterrows():
        pred_text_embedding = row['text_embedding']
        best_match_sct_name = None
        highest_similarity = -1 # Initialize with a value lower than any possible cosine similarity

        for merged_df in merged_dfs:
            # Calculate similarity between the pred text embedding and all merged text embeddings in the current merged_df
            similarity_scores = calculate_similarity(pred_text_embedding, list(merged_df['value_y_embedding']))

            # Find the index of the highest similarity score in the current merged_df
            best_match_index_in_merged = similarity_scores.argmax()
            current_highest_similarity = similarity_scores[best_match_index_in_merged]

            # Check if this is the best match found so far across all merged_dfs and if it's above the threshold Using a threshold of 0.7
            if current_highest_similarity > 0.7 and current_highest_similarity > highest_similarity:
                highest_similarity = current_highest_similarity
                best_match_sct_name = merged_df.loc[best_match_index_in_merged, 'sct_name']

        sct_names_for_df.append(best_match_sct_name)
        similarity_scores_for_df.append(highest_similarity)

    pred_df['sct_name_semantic'] = sct_names_for_df
    pred_df['similarity_score'] = similarity_scores_for_df
    all_updated_pred_dfs.append(pred_df)

# Concatenate the updated pred_text dataframes
combined_pred_text_df_semantic = pd.concat(all_updated_pred_dfs, ignore_index=True)

# Display the final dataframe
display(combined_pred_text_df_semantic)
```

## --- 5. evaluating the string match approach ---

```python
import pandas as pd

# combined_pred_text_df is the result from the fuzzy matching approach

# Calculate the number of rows where a match was found
matched_rows_count_fuzzy = combined_pred_text_df['sct_name'].notna().sum()

# Calculate the total number of rows in the dataframe
total_rows_count = len(combined_pred_text_df)

# Calculate the custom accuracy
custom_accuracy_fuzzy = matched_rows_count_fuzzy / total_rows_count if total_rows_count > 0 else 0

print(f"Number of rows with a fuzzy match: {matched_rows_count_fuzzy}")
print(f"Total number of rows: {total_rows_count}")
print(f"Custom accuracy for fuzzy matching: {custom_accuracy_fuzzy:.4f}")
```

```python
# combined_pred_text_df_semantic is the result from the semantic matching approach

# Calculate the number of rows where a match was found (sct_name_semantic is not None)
matched_rows_count_semantic = combined_pred_text_df_semantic['sct_name_semantic'].notna().sum()

# Calculate the total number of rows in the dataframe (should be the same as for fuzzy)
total_rows_count_semantic = len(combined_pred_text_df_semantic)

# Calculate the custom accuracy
custom_accuracy_semantic = matched_rows_count_semantic / total_rows_count_semantic if total_rows_count_semantic > 0 else 0

print(f"Number of rows with a semantic match: {matched_rows_count_semantic}")
print(f"Total number of rows: {total_rows_count_semantic}")
print(f"Custom accuracy for semantic matching: {custom_accuracy_semantic:.4f}")
```
